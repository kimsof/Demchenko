```python
import numpy as np
import pandas as ps

def ansol(r, phi, t):   # вывод аналитического решения
    array = np.zeros((len(phi), len(r)))
    for m in range(len(phi)):
        for l in range(len(r)):
            array[m][l] = r[l]*np.cos(phi[m])/np.sqrt(5 - 4*t)
    return array

def table(array, x=None, y=None):                 # для вывода таблицы с ответом
    return ps.DataFrame(array, y, x)
```

```python
mu = 2
tau = 0.1
h_r = 0.1
h_phi = np.pi/20
N = int(1/tau)
L = int(1/h_r)
M = int(np.pi/2/h_phi)
t = np.linspace(0, 1, N + 1)
r = np.linspace(0, 1, L + 1)
phi = np.linspace(0, np.pi/2, M + 1)
eps = 10**-2
u_analytical = ansol(r, phi, 0)
u_n = u_analytical
k = 0

for n in range(N):
    k = 0
    print('n = ' + str(n))
    u0 = u_n                        # u_k
    u1 = np.ones(u0.shape)          # u_tilda
    u2 = np.ones(u0.shape)          # u_k+1
    if n == 0:
        print('u0:')
        print(table(u0))
        print('u2:')
        print(table(u2))
        
    condition = 1
    
    while (condition > eps):
        print('k = ' + str(k))
        #u_n = u                   # u_n
        #u0 = u_n                  # u_k
        #u1 = np.zeros(u0.shape)   # u_tilda
        #u2 = np.zeros(u0.shape)   # u_k+1

        # первая прогонка
        u1[0, :] = r/np.sqrt(5 - 4*t[n + 1])
        u1[M, :] = 0
        r = np.linspace(0, 1, L*2 + 1)
        for m in range(1, M):
            
            # заполнение a, b, c, d
            a = np.zeros(L + 1)
            b = np.zeros(L + 1)
            c = np.zeros(L + 1)
            d = np.zeros(L + 1)

            b[0] = 1  # a[0] = 0, c[0] = 0
            d[0] = 0  # = u[0][m]
            a[L] = 1  # b[L] = 0
            d[L] = np.cos(phi[m])/np.sqrt(5 - 4*t[n+1]) 
            
            for l in range(1, L):
                a[l] = -r[2*l + 1]*(u0[m][l + 1]**mu + u0[m][l]**mu)*tau/(2*r[2*l]*h_r**2)
                c[l] = -r[2*l - 1]*(u0[m][l]**mu + u0[m][l - 1]**mu)*tau/(2*r[2*l]*h_r**2)
                b[l] = 1 - a[l] - c[l]
                d[l] = u_n[m][l]          
            
            # прямая прогонка
            alpha = np.zeros(L)
            beta = np.zeros(L)
            alpha[0] = -c[0]/b[0]
            beta[0] = d[0]/b[0]
            for i in range(1, L):
                alpha[i] = -c[i]/(b[i] + a[i]*alpha[i - 1])
                beta[i] = (d[i] - a[i]*beta[i - 1])/(b[i] + a[i]*alpha[i - 1])

            # обратная прогонка
            u1[m][L] = (d[L] - a[L]*beta[L - 1])/(alpha[L - 1]*a[L] + b[L])
            for l in range(-(L - 1), 1):
                u1[m][int(np.fabs(l))] = alpha[int(np.fabs(l))]*u1[m][int(np.fabs(l)) + 1] + beta[int(np.fabs(l))]
                
                
        # вторая прогонка
        u2[:, 0] = 0
        u2[:, L] = np.cos(phi)/np.sqrt(5 - 4*t[n + 1])
        for l in range(1, L):
            
            # заполнение a, b, c, d
            a = np.zeros(M + 1)
            b = np.zeros(M + 1)
            c = np.zeros(M + 1)
            d = np.zeros(M + 1) 

            b[0] = 1  
            d[0] = r[2*l]/np.sqrt(5 - 4*t[n+1])
            a[M] = 1
            d[M] = 0
            
            for m in range(1, M):
                a[m] = -(u0[m + 1][l]**mu + u0[m][l]**mu)*tau/(2*r[2*l]**2*h_phi**2)
                c[m] = -(u0[m][l]**mu + u0[m - 1][l]**mu)*tau/(2*r[2*l]**2*h_phi**2)
                b[m] = 1 - a[m] - c[m]
                d[m] = u1[m][l]    

            # прямая прогонка
            alpha = np.zeros(M)
            beta = np.zeros(M)
            alpha[0] = -c[0]/b[0]
            beta[0] = d[0]/b[0]
            for i in range(1, M):
                alpha[i] = -c[i]/(b[i] + a[i]*alpha[i - 1])
                beta[i] = (d[i] - a[i]*beta[i - 1])/(b[i] + a[i]*alpha[i - 1])

            # обратная прогонка
            u2[M][l] = (d[M] - a[M]*beta[M - 1])/(alpha[M - 1]*a[M] + b[M])
            for m in range(-(M - 1), 1):
                u2[int(np.fabs(m))][l] = alpha[int(np.fabs(m))]*u2[int(np.fabs(m)) + 1][l] + beta[int(np.fabs(m))]
            
        r = r = np.linspace(0, 1, L + 1)
        k += 1
        print('u0:')
        print(table(u0))
        print('u2:')
        print(table(u2))
        condition = max(map(max, np.fabs((u2[1 : u2.shape[0] - 1, 1 : u2.shape[1] - 1] - u0[1 : u0.shape[0] - 1, 1 : u0.shape[1] - 1])/u2[1 : u2.shape[0] - 1, 1 : u2.shape[1] - 1])))
        u0 = u2
    u_n = u2
```
