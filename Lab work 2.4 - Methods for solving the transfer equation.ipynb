```python
import numpy as np
import pandas as ps
L = 1

def a_coef(x):
    return x + 4

def psi(t):
    return 16*np.exp(-2*t)

def psi1(t):
    return -32*np.exp(-2*t)

def psi2(t):
    return 64*np.exp(-2*t)

def psi3(t):
    return -128*np.exp(-2*t)

def analytical(x, t):
    return np.exp(-2*(t - np.log(x + 4)))

def p(array):
    return ps.DataFrame(array)
#return ps.DataFrame(array)[2*np.linspace(0, 10, 11)]

def utx(array, x, t):
    return ps.DataFrame(array, t, x)

def numerical(N):
    h = L/(N-1)
    x = np.linspace(0, L, N)

    t = [0]
    tau = h/4
    
    a = a_coef(x)  
    a0 = a[0]

    #заполнение сетки  
    u = (x + 4)**2   # u(x, t=0)
    
    while t[len(t) - 1] <= 1:
    
        n = len(t) - 1
        u = np.vstack((u, np.zeros(len(x))))       # добавление n+1 слоя по t
        t.append(t[n] + tau)  # инкрементирование по t

        # чтобы получить значения u[n+1][1], u[n+1][2] нужны u'[n+1][0], u"[n+1][0], u'''[n+1][0]
        u1 = -psi1(t[n+1])/a0                              # u'[n+1][0]
        u2 = (psi2(t[n+1]) + psi1(t[n+1]))/(a0**2)         # u"[n+1][0]
        u3 = -psi3(t[n+1])/(a0**3) - 3*u2/a0 - u1/(a0**2)  # u'''[n+1][0]

        # заполнение n+1 слоя
        u[n+1][0] = psi(t[n+1])
        u[n+1][1] = u[n+1][0] + u1*h + u2*h**2/2 + u3*h**3/6
        u[n+1][2] = u[n+1][0] + u1*2*h + u2*2*h**2 + u3*4*h**3/3
        for m in range(3, len(x)):  
            u[n+1][m] = u[n][m] + (tau/(6*h))*(x[m] + 4)*(1 - tau/2 + tau**2/6)*(2*u[n][m-3] - 9*u[n][m-2] + 18*u[n][m-1] - 11*u[n][m]) + (tau**2/(2*h**2))*(1 - tau)*(-u[n][m-3] + 4*u[n][m-2] - 5*u[n][m-1] + 2*u[n][m])*(x[m] + 4)**2 - (tau**3/(6*h**3))*(-u[n][m-3] + 3*u[n][m-2] - 3*u[n][m-1] + u[n][m])*(x[m] + 4)**3
    x_output = np.linspace(0, 1, 11)
    u_output = np.zeros(11)
    u_M = u[int(round(1/tau))]
    for i in range(11):
        u_output[i] = u_M[int(round(x_output[i]/h))]
    return u_output

def output(N):
    print('число узлов: ' + str(N))
    x_output = np.linspace(0, 1, 11)
    array = np.zeros((11, 4))
    array.T[0] = x_output
    array.T[1] = analytical(x_output, 1)
    array.T[2] = numerical(N)
    array.T[3] = np.fabs(array.T[2] - array.T[1])
    columns = ['x', 'analytical', 'numerical', 'rate']
    return ps.DataFrame(array, x_output, columns)
```
