## Лабораторная работа №5
### Задача: 
### $ \left\{\begin{matrix}
\frac{\partial u}{\partial t} = \frac{\partial}{r \partial r}(ru^{2} \frac{\partial u}{\partial r}) + \frac{\partial }{r^{2}\partial \varphi} (u^{2} \frac{\partial u}{\partial \varphi }), \\
t>0, \hspace{5pt} r\leq 1, \hspace{5pt} 0<\varphi <\frac{\pi }{2}; \\
u(0, r, \varphi) = r cos \varphi\frac{1}{\sqrt{5}}, \hspace{5pt} 0\leq r\leq 1, \hspace{5pt} 0\leq \varphi \leq \frac{\pi}{2}; \\
u(t, 0, \varphi) = 0, \hspace{5pt} 0 < t\leq 1, \hspace{5pt} 0\leq \varphi \leq \frac{\pi}{2}; \\
u(t, 1, \varphi) = cos \varphi\frac{1}{\sqrt{5 - 4t}}, \hspace{5pt} 0< t \leq 1, \hspace{5pt} 0\leq \varphi \leq \frac{\pi}{2}; \\
u(t, r, 0) = r \frac{1}{\sqrt{5 - 4t}}, \hspace{5pt} 0 < t \leq 1, \hspace{5pt} 0 < r < 1; \\
u(t, r, \frac{\pi }{2}) = 0, \hspace{5pt} 0 < t \leq 1, \hspace{5pt} 0 < r < 1; \\
\end{matrix}\right. $
    
### Аналитическое решение: 
### $ u = r cos \varphi \frac{1}{\sqrt{5 - 4t} }$

### Сетка:
### $ D_h = \begin{Bmatrix} 
(r_l, \varphi _m, t^n): r_l = h_r l, l = \overline{0, L}, h_r L = 1; \\ \varphi _m = h_\varphi  m, l = \overline{0, M}, h_\varphi  M = \frac{\pi }{2}; \\ t^n = n\tau, n = \overline{0, N}, \tau  N = \frac{5}{4} - \Delta , \Delta = const;
\end{Bmatrix}  $
    
### Разностная схема: 
### $ \left\{\begin{matrix}
\frac{\tilde{u}_{l,m} - u^{n}_{l,m}}{\tau} = \frac{r_{l+\frac{1}{2}}[(u^k_{l+1,m})^\mu + (u^k_{l,m})^\mu](\tilde{u}_{l+1,m} - \tilde{u}_{l,m})}{2r_lh^2_r} - \frac{r_{l-\frac{1}{2}}[(u^k_{l,m})^\mu + (u^k_{l-1,m})^\mu](\tilde{u}_{l,m} - \tilde{u}_{l-1,m})}{2r_lh^2_r}, \hspace{5pt} l = \overline{1, L - 1}, \hspace{5pt} m = \overline{1, M - 1} \\
\tilde{u}_{0,m} = 0, \hspace{5pt} \tilde{u}_{L,m} = cos \varphi _m\frac{1}{\sqrt{5-4t^{n+1}}}, \hspace{5pt} m = \overline{0, M} \\
\end{matrix}\right. $
### $ \left\{\begin{matrix}
\frac{u^{k+1}_{l,m} - \tilde{u}_{l,m}}{\tau} = \frac{[(u^k_{l,m+1})^\mu + (u^k_{l,m})^\mu](u^{k+1}_{l,m+1} - u^{k+1}_{l,m})}{2r_l^2 h^2_r} - \frac{[(u^k_{l,m})^\mu + (u^k_{l,m-1})^\mu](u^{k+1}_{l,m} - u^{k+1}_{l,m-1})}{2r^2_l h^2_r}, \hspace{5pt} l = \overline{1, L - 1}, \hspace{5pt} m = \overline{1, M - 1} \\ 
u^{k+1}_{l,0} = r_l \frac{1}{\sqrt{5 - 4t^{n+1}}}, \hspace{5pt} u^{k+1}_{l,M} = 0, \hspace{5pt} l = \overline{0, L - 1}, \\
k = \overline{n, J}, J = min(k): \displaystyle \max_{l,m} \left| \frac{u^{k+1}_{l,m} - u^k_{l,m}}{u^{k+1}_{l,m}}  \right| \leq \varepsilon = 10^{-4}, \\
u^{n+1}_{l,m} = u^{J+1}_{l,m}, \hspace{5pt} l = \overline{1, L-1}, \hspace{5pt} m = \overline{1, M-1}, \hspace{5pt} n = \overline{0, N-1}, \\
u^0_{l,m} = r cos\varphi _m, \hspace{5pt} l = \overline{0,L}, \hspace{5pt} m = \overline{0, M}
\end{matrix}\right.$

```python
import numpy as np
import pandas as ps

def ansol(r, phi, t):   # вывод аналитического решения
    array = np.zeros((len(phi), len(r)))
    for m in range(len(phi)):
        for l in range(len(r)):
            array[m][l] = r[l]*np.cos(phi[m])/np.sqrt(5 - 4*t)
    return array

def table(array, x=None, y=None):                 # для вывода таблицы с ответом
    return ps.DataFrame(array, y, x)
```

```python
def solver(L, M, N):
    
    print('L = ' + str(L), ', M = ' + str(M), ', N = ' + str(N))
    
    mu = 2
    Ct = 5
    epsilon = 10**-4
    T = 1
    
    h_r = 1/L              # 0 <=  r  <= 1
    h_phi = np.pi/2 / M    # 0 <= phi <= pi/2
    tau = T/N              # 0 <=  t  <= T
    
    r = np.linspace(0, 1, L + 1)
    phi = np.linspace(0, np.pi/2, M + 1)
    t = np.linspace(0, T, N + 1)
    
    u_n = ansol(r, phi, 0)  # u при t = 0
    
    for n in range(N):

        u_k = u_n                         # u_k
        u_tilda = np.ones(u_k.shape)      # u_tilda
        u_k_plus_1 = np.ones(u_k.shape)   # u_k+1

        condition = 1

        while (condition > epsilon):
            
            # первая прогонка

            u_tilda[0, :] = r/np.sqrt(5 - 4*t[n + 1])
            u_tilda[M, :] = 0

            r = np.linspace(0, 1, L*2 + 1) # удвоение для подсчета коэфов

            for m in range(1, M):

                # заполнение a, b, c, d

                a = np.zeros(L + 1)
                b = np.zeros(L + 1)
                c = np.zeros(L + 1)
                d = np.zeros(L + 1)

                b[0] = 1      
                d[0] = 0      
                b[L] = 1      
                d[L] = np.cos(phi[m])/np.sqrt(5 - 4*t[n + 1]) 

                for l in range(1, L):
                    c[l] = -r[2*l + 1]*(u_k[m, l + 1]**mu + u_k[m, l]**mu)*tau/(2*r[2*l]*h_r**2)
                    a[l] = -r[2*l - 1]*(u_k[m, l]**mu + u_k[m, l - 1]**mu)*tau/(2*r[2*l]*h_r**2)
                    b[l] = 1 - a[l] - c[l]
                    d[l] = u_n[m, l]          

                # прямая прогонка

                alpha = np.zeros(L)
                beta = np.zeros(L)

                alpha[0] = -c[0]/b[0]
                beta[0] = d[0]/b[0]

                for i in range(1, L):
                    alpha[i] = -c[i]/(b[i] + a[i]*alpha[i - 1])
                    beta[i] = (d[i] - a[i]*beta[i - 1])/(b[i] + a[i]*alpha[i - 1])

                # обратная прогонка

                u_tilda[m, L] = (d[L] - a[L]*beta[L - 1])/(alpha[L - 1]*a[L] + b[L])

                for l in range(-(L - 1), 1):
                    u_tilda[m, int(np.fabs(l))] = alpha[int(np.fabs(l))]*u_tilda[m, int(np.fabs(l)) + 1] + beta[int(np.fabs(l))]


            # вторая прогонка

            u_k_plus_1[:, 0] = 0
            u_k_plus_1[:, L] = np.cos(phi)/np.sqrt(5 - 4*t[n + 1])

            for l in range(1, L):

                # заполнение a, b, c, d

                a = np.zeros(M + 1)
                b = np.zeros(M + 1)
                c = np.zeros(M + 1)
                d = np.zeros(M + 1) 

                b[0] = 1  
                d[0] = r[2*l]/np.sqrt(5 - 4*t[n+1])
                b[M] = 1
                d[M] = 0

                for m in range(1, M):
                    c[m] = -(u_k[m + 1, l]**mu + u_k[m, l]**mu)*tau/(2*r[2*l]**2*h_phi**2)
                    a[m] = -(u_k[m, l]**mu + u_k[m - 1, l]**mu)*tau/(2*r[2*l]**2*h_phi**2)
                    b[m] = 1 - a[m] - c[m]
                    d[m] = u_tilda[m, l]    

                # прямая прогонка

                alpha = np.zeros(M)
                beta = np.zeros(M)

                alpha[0] = -c[0]/b[0]
                beta[0] = d[0]/b[0]

                for i in range(1, M):
                    alpha[i] = -c[i]/(b[i] + a[i]*alpha[i - 1])
                    beta[i] = (d[i] - a[i]*beta[i - 1])/(b[i] + a[i]*alpha[i - 1])

                # обратная прогонка

                u_k_plus_1[M, l] = (d[M] - a[M]*beta[M - 1])/(alpha[M - 1]*a[M] + b[M])

                for m in range(-(M - 1), 1):
                    u_k_plus_1[int(np.fabs(m)), l] = alpha[int(np.fabs(m))]*u_k_plus_1[int(np.fabs(m)) + 1, l] + beta[int(np.fabs(m))]

            r = np.linspace(0, 1, L + 1)
            
            condition = max(map(max, np.fabs((u_k_plus_1[1 : u_k_plus_1.shape[0] - 1, 1 : u_k_plus_1.shape[1] - 1] - u_k[1 : u_k.shape[0] - 1, 1 : u_k.shape[1] - 1])/u_k_plus_1[1 : u_k_plus_1.shape[0] - 1, 1 : u_k_plus_1.shape[1] - 1])))
            u_k = u_k_plus_1
        u_n = u_k_plus_1
        
    u_analytical = ansol(r, phi, T)
    u_numerical = u_n
    substract = np.fabs(u_analytical - u_numerical)
    
    df_analytical = table(u_analytical, r, phi)
    df_analytical = df_analytical.iloc[::int(np.pi/2/5/h_phi), ::int(0.2/h_r)]
    df_numerical = table(u_numerical, r, phi)
    df_numerical = df_numerical.iloc[::int(np.pi/2/5/h_phi), ::int(0.2/h_r)]
    df_substract = table(substract, r, phi)
    df_substract = df_substract.iloc[::int(np.pi/2/5/h_phi), ::int(0.2/h_r)]
    
    return df_analytical, df_numerical, df_substract
```
