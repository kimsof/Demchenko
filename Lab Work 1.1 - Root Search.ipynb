Локализация корней

```python
# по следствию из основной теоремы алгебры
a=abs(a6)/(abs(a6)+a2)
b=1+a2/a0
print('Все корни уравнения расположены на отрезке\n[' + str(a) + ',' + str(b) + ']\nВсего либо 1, либо 3 положительных действительных корня')
```

```python
# функция и ее производные
def f(Y):
    return a0*Y**6+a1*Y**5+a2*Y**4+a3*Y**3+a4*Y**2+a5*Y+a6
def f1(Y):
    return 6*a0*Y**5+5*a1*Y**4+4*a2*Y**3+3*a3*Y**2+2*a4*Y+a5
def f2(Y):
    return 5*6*a0*Y**4+4*5*a1*Y**3+3*4*a2*Y**2+2*3*a3*Y+2*a4
def f3(Y):
    return 4*5*6*a0*Y**3+3*4*5*a1*Y**2+2*3*4*a2*Y+2*3*a3
def f4(Y):
    return 3*4*5*6*a0*Y**2+2*3*4*5*a1*Y+2*3*4*a2
def f5(Y):
    return 2*3*4*5*6*a0*Y+2*3*4*5*a1
f6 = 2*3*4*5*6*a0
```

```python
# вспомогательные функции для Th Бюдана-Фурье
def aux_str(b,f):
    array = np.zeros((1,7))
    array[0][:] = [f(b),f1(b),f2(b),f3(b),f4(b),f5(b),f6]
    return array

def aux_point(q,a,b):
    return a + (b-a)/(2**q)
    
def BF(w,a,b):
    a=(abs(a6))/(abs(a6)+a2)
    array = np.zeros((w+1,7))
    array[0][:] = aux_str(b,f)
    i=1
    while i < w+1:
        array[i][:] = aux_str(aux_point(i,a,b),f)
        i+=1
    table = ps.DataFrame(data=array, index=np.arange(0,w+1), columns=np.arange(0,7))
    table.columns = ['f','f1','f2','f3','f4','f5','f6']
    return table
```

```python
# ряд Штурма
def F(Y):
    return a0*Y**6+a1*Y**5+a2*Y**4+a3*Y**3+a4*Y**2+a5*Y+a6
def F1(Y):
    return 6*a0*Y**5+5*a1*Y**4+4*a2*Y**3+3*a3*Y**2+2*a4*Y+a5
def F2(Y):
    return -(F(Y) % F1(Y))
def F3(Y):
    return -(F1(Y) % F2(Y))
def F4(Y):
    return -(F2(Y) % F3(Y))
def F5(Y):
    return -(F3(Y) % F4(Y))
def F6(Y):
    return -(F4(Y) % F5(Y))
```

```python
def SH(w,a,b):
    array = np.zeros((w+1,7))
    array[0][:] = aux_str(b,F)
    i=1
    while i < w+1:
        array[i][:] = aux_str(aux_point(i,a,b),F)
        i+=1
    table = ps.DataFrame(data=array, index=np.arange(0,w+1), columns=np.arange(0,7))
    table.columns = ['F','F1','F2','F3','F4','F5','F6']
    return table
```

Нахождение локализованных корней методом Ньютона

```python
w11 = aux_point(4,a,b)
w12 = aux_point(3,a,b)
w21 = aux_point(10,a,b)
w22 = aux_point(9,a,b)
w31 = aux_point(19,a,b)
w32 = aux_point(18,a,b)
# расстояние от w11 до 1 корня больше, чем расстояние до 2 корня => чтобы не было ошибок сократим отрезок локализации первого корня
dot = aux_point(1,w11,w12)
w11 = dot
w11
```

```python
# функция, выводящая начальное приблежение на заданном отрезке
def search_point(a,b):
    q = 1
    x0 = aux_point(q,a,b)
    while f(x0)*f2(x0)<0:
        q +=1
        x0 = aux_point(q,a,b)
    return x0
```

```python
def Newton_method(E,w1,w2):
    s = search_point(w1,w2)  
    t = s - f(s)/f1(s)
    M2 = f2(max(t,s))
    m1 = f1(min(t,s))
    while (M2*(abs(t-s))**2)/(2*m1) > E:
        s = t
        t = t - f(t)/f1(t)
        M2 = f2(max(t,s))
        m1 = f1(min(t,s))
    return t

def answer(E):
    answer_array = [Newton_method(E,w11,w12),Newton_method(E,w21,w22),Newton_method(E,w31,w32)]
    table3 = ps.DataFrame(data=answer_array, index=np.arange(0,3), columns=np.arange(0,1))
    table3.index = ['Y1','Y2','Y3']
    return table3
```
