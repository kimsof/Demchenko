```python
import numpy as np
import pandas as ps

def p(array):
    return ps.DataFrame(array)

def anal(x, y):
    array = np.zeros((len(y), len(x)))
    for i in range(len(x)):
        for j in range(len(y)):
            array[j][i] = x[i]*(1-x[i])*y[j]*(1-y[j])
    return array

def error(x, y, u):
    du = anal(x,y) - u
    return np.linalg.norm(du.T, 1)/np.linalg.norm(u.T, 1)
    
```
